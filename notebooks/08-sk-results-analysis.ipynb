## LLAMA

```python
fig, axs = plt.subplots(1,4,sharey=True, figsize=(10,3))
for n, col in enumerate(cbd_columns[:4]):
    axs[n].bar(llama321005_results.index, llama321005_results[col])
    axs[n].set_title(col.split("_")[-1])
    #plt.savefig(RESULTS_DIR/('llama3.2-1B_' + col+'.png'))
    #plt.show()
fig.suptitle('Llama 3.2 1B T=0.5')
plt.tight_layout()
```
